```python
import cv2
import scipy
import skimage
import scipy.io
import numpy as np
import scipy.ndimage
from glob import glob
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
```

```python
mat_paths = glob("/content/*.mat")
```

# Gaussian_filter

Gaussian filter always produses different results

```python
for i in range(len(mat_paths)):
    mat = scipy.io.loadmat(mat_paths[i])
    foggy_image = mat['foggy_image']
    Gauss_img = mat['Gauss_img']
    Gray_img = mat['Gray_img']
    Py_Gauss = skimage.filters.gaussian(foggy_image, sigma=10, mode = 'nearest', truncate=2.0)
    print(np.max(Gauss_img-Py_Gauss))
    plt.imshow(Gauss_img)
    plt.show()
    plt.imshow(Py_Gauss)
    plt.show()
    # plt.imshow(Gauss_img-Py_Gauss)
    # plt.show()
```

# Gray scale Image

matlab and cv2rgb2gray are producing same results

```python
for i in range(len(mat_paths)):
    mat = scipy.io.loadmat(mat_paths[i])
    foggy_image = mat['foggy_image']
    Gauss_img = mat['Gauss_img']
    Gray_img = mat['Gray_img']
    Py_Gray1 = rgb2gray(foggy_image/255)
    Py_Gray = cv2.cvtColor(foggy_image, cv2.COLOR_RGB2GRAY)/255
    print('CV2 RGB2GRAY : ',np.max(Py_Gray-Gray_img),'skimage RGB2GRAY : ',np.max(Py_Gray1-Gray_img))
    plt.imshow(Gray_img,'gray')
    plt.show()
    plt.imshow(Py_Gray1,'gray')
    plt.show()
    plt.imshow(Py_Gray,'gray')
    plt.show()
```
